```python
for column in til_ped.columns:
    print('The column ' + column + ' has ' + str(til_ped[column].nunique()) + ' unique records')
```

Change the name of the columns to `Id`, `Sire`, `Dam` and `Population`.

```python
til_ped.columns = ['Id', 'Sire', 'Dam', 'Population']
```

```python
til_ped
```

For each Sire get the number of offspring.

```python
til_ped['Sire'].value_counts()[til_ped['Sire'].unique()]
```

```python
for sire in til_ped['Sire'].unique():
    print(sire + ' has ' + str(til_ped[til_ped['Sire']==sire]['Id'].nunique()) + ' offspring')
```

For each family (sire and dam) get the number of offspring.

```python
til_ped['Family'] = til_ped['Sire'] + '_' + til_ped['Dam']
```

```python
family_labels = ["Family_0" + str(i) for i in range(1,37)]
```

```python
original_labels = list(til_ped["Family"].unique())
```

```python
family_labels = dict(zip(original_labels, family_labels))
```

```python
til_ped["Family"] = til_ped["Family"].map(family_labels)
```

```python
til_ped
```

```python
til_ped['Family'].value_counts()[til_ped['Family'].unique()]
```

```python
for family in til_ped['Family'].unique():
    print(family + ' has ' + str(til_ped[til_ped['Family']==family]['Id'].nunique()) + ' animals')
```

Assign an index that should contain the string `til_` followed by the number of the row. E.g. the index of the first recording should be `til_1` etc.

```python
til_ped.index = ['til_' + str(i) for i in til_ped.index]
til_ped
```

Extract the 10th row of the dataframe.

```python
til_ped.iloc[9]
```

Create a new dataframe that contains the values of the original dataframe located between the 200th-300th rows and on the first three columns.

```python
til_ped_new = til_ped.iloc[199:301,:3]
til_ped_new
```

Extract the rows where the last column contains values of either `Mindu` or `Kilola`.

```python
til_ped_sub1 = til_ped[(til_ped['Population']=='Mindu')|(til_ped['Population']=='Kilola')]
til_ped_sub1
print(set(til_ped_sub1['Population']))
```

Extract the rows where the last column doesn't contail values of `Ruaha` and `Utete`.

```python
til_ped_sub2 = til_ped[(til_ped['Population'] != 'Ruaha') & (til_ped['Population'] != 'Utete')]
til_ped_sub2
print(set(til_ped_sub2['Population']))
```

How many records are in the dataframe.

For the non-numeric columns estimate the number of unique records.

```python
non_numeric = ['Animal_Id', 'Location', 'Line']
for column in non_numeric:
    print(column + ' has ' + str(til_pheno[column].nunique()) + ' values')
```

Show a descriptive statistics summary.

```python
til_pheno.describe()
```

Which `Location` has the animals with the highest mean final weight.

```python
set(til_pheno['Location'])
```

```python
sum(til_pheno[til_pheno['Location']=='Kunduchi']['Weight_final'])/len(til_pheno[til_pheno['Location']=='Kunduchi']['Weight_final'])
```

```python
til_pheno[til_pheno['Location']=='Kunduchi']['Weight_final'].mean()
```

```python
til_pheno[til_pheno['Location']=='Pangani']['Weight_final'].mean()
```

Which `Line` has the lowest mean final weight.

```python
lines = set(til_pheno['Line'])
lines
```

```python
for line in lines:
    mean_line = til_pheno[til_pheno['Line']==line]['Weight_final'].mean()
    print(line + ' has a mean final weight of ' + str(mean_line))
```

Estimate correlations between the weight and length related columns.

```python
til_pheno[['Weight_initial','Weight_final','Length_initial','Length_final']].corr()
```

Create a new dataframe keeping only records where the difference between `Weight_final` and `Weight_initial` is above 100.

```python
til_pheno_red1 = til_pheno[(til_pheno['Weight_final'] - til_pheno['Weight_initial']) > 100]
til_pheno_red1
```

Create a new dataframe keeping only records where the difference between `Weight_final` and `Weight_initial` is above 120 and the difference between `Length_final` and `Length_initial` is above 5.

```python
til_pheno_red2 = til_pheno[(til_pheno['Weight_final'] - til_pheno['Weight_initial'] > 120) 
                           & (til_pheno['Length_final'] - til_pheno['Length_initial'] > 5)]
til_pheno_red2
```

Create histograms depicting the `Weight_final` for each of the `Location.

```python
set(til_pheno['Location'])
```

```python
til_pheno[til_pheno['Location']=='Kunduchi']['Weight_final'].plot(kind='hist',bins=50)
```

```python
til_pheno[til_pheno['Location']=='Pangani']['Weight_final'].plot(kind='hist',bins=50)
```

Create a boxplot depicting the difference between `Weight_final` and `Weight_initial` for each `Line`.

```python
til_pheno['Weight_diff'] = til_pheno['Weight_final'] - til_pheno['Weight_initial']
til_pheno.boxplot(column='Weight_diff', by='Line')
```
