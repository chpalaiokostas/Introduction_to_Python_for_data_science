```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
diamonds.info()
```

```python
plt.style.use('seaborn-v0_8-colorblind')
```

Create a scatter plot with the `carat` on the x axis and the `price` variable on the y axis.

```python
fig, ax = plt.subplots()
ax.scatter(diamonds['carat'],diamonds['price'])
ax.set_xlabel('Carats')
ax.set_ylabel('Price')
```

Create the same scatter plot as above, but this type use a different colour for each `cut` value.

```python
colours = ['blue','cyan', 'purple', 'green', 'orange']
cut_options = diamonds['cut'].unique()
cut_colours = list(zip(cut_options,colours))
cut_colours
```

```python
fig, ax = plt.subplots()
for value, colour in cut_colours:
    ax.scatter(diamonds[diamonds['cut']==value]['carat'], diamonds[diamonds['cut']==value]['price'],c=colour, alpha=0.5,label=value)
ax.set_xlabel('Carats')
ax.set_ylabel('Price')
ax.legend()
```

Create a bar chart showing the price for the different colours. Add also standard error bars. To calculate the standard error of the mean you can use the `scipy.stats.sem`.

```python
from scipy.stats import sem
diamonds_summary = diamonds.groupby('color')['price'].agg([np.mean,sem])
diamonds_summary 
```

```python
fig, ax = plt.subplots()
ax.bar(diamonds_summary.index, diamonds_summary['mean'], yerr=diamonds_summary['sem'])
ax.set_xlabel('Colour')
ax.set_ylabel('Price')
```

Create a stacked bar chart including the mean values of the `x`, `y` and `z` columns for each clarity value.

```python
clarity_summary = diamonds.groupby('clarity')[['x','y','z']].mean()
clarity_summary
```

```python
fig, ax = plt.subplots()
ax.bar(clarity_summary.index, clarity_summary['x'], label='x')
ax.bar(clarity_summary.index, clarity_summary['y'], bottom=clarity_summary['x'], label='y')
ax.bar(clarity_summary.index, clarity_summary['z'], 
       bottom=clarity_summary['x'] + clarity_summary['y'], label='z')
ax.set_xlabel('Clarity')
ax.set_ylabel('Size')
ax.legend()
```

Create a boxplot with the price range for the different `cut` values.

```python
cut_options
```

```python
diamonds_ideal = diamonds[diamonds['cut']=='Ideal']['price']
diamonds_premium = diamonds[diamonds['cut']=='Premium']['price']
diamonds_good = diamonds[diamonds['cut']=='Good']['price']
diamonds_very_good = diamonds[diamonds['cut']=='Very Good']['price']
diamonds_fair = diamonds[diamonds['cut']=='Fair']['price']
```

```python
diamonds_ideal
```

```python
fig, ax = plt.subplots()
ax.boxplot([diamonds_fair,diamonds_good,diamonds_very_good,diamonds_premium,diamonds_ideal])
ax.set_xticklabels(['Fair','Good','Very Good','Premium','Ideal'])
ax.set_xlabel('Cut')
ax.set_ylabel('Price')
```

Calculate the Pearson correlations amongst all numeric columns. Create a heatmap plot from the correlation matrix, including a colour bar showing the range of colours for the corresponding correlation values. 

```python
diamonds.info()
```

```python
corr_mat = diamonds.loc[:,["carat","depth","table","price","x","y","z"]].corr()
corr_mat
```

```python
labels = ["carat", "depth", "table", "price", "x", "y", "z"]
fig, ax = plt.subplots()
diamonds_heatmap = ax.imshow(corr_mat)
plt.colorbar(diamonds_heatmap)
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels,fontsize=16, rotation=45)
plt.yticks(tick_marks, labels, fontsize=16)
```

### Exercise 2

We will work with the `snp_variants.txt` dataset. For all the plots use a customed style of your choice.


* Create a horizontal bar chart with the number of SNP variants of each chromosome.
* Create separate line plots for each chromosome depicting the base composition of the `Reference` column along non-overlapping windows of 10,000,000 base pairs (you will need to estimate first the frequency of each base [A,G,C,T] for every ten million base pairs). Save the plot in a pdf format.
* In the case of the first chromosome include annotation text. In particular add to the plot the text `Position of interest` at a suitable location and point using an arrow to the location where the frequency of the G base is at it's lowest.

```python
snp_variants = pd.read_csv('snp_variants.txt', sep='\t')
snp_variants.info()
```

```python
snp_variants
```

Create a horizontal bar chart with the number of SNP variants of each chromosome.

```python
snp_variants.groupby('Chromosome').size().plot(kind='barh')
```

Create separate line plots for each chromosome depicting the base composition of the `Reference` column along non-overlapping windows of 10,000,000 base pairs (you will need to estimate the occurences each base [A,G,C,T] for every ten million base pairs). Save the plot in a pdf format.

```python
chromosome = snp_variants['Chromosome'].unique()
chromosome
```

```python
snp_first_chrom = snp_variants[snp_variants['Chromosome']==chromosome[0]]
bases_chr1 = {'A':[0], 'G':[0], 'C':[0], 'T':[0]}
for i in range(0,snp_first_chrom['Position'].max(), 10000000):
    win_variants = snp_first_chrom[(snp_first_chrom['Position'] >= i) 
                                   & (snp_first_chrom['Position'] <= i + 10000000)]['Reference']
    for base in 'AGCT':
        bases_chr1[base].append(win_variants.value_counts()[base]/win_variants.value_counts().sum())
```

```python
chrom1_regions = [i for i in range(0,snp_first_chrom['Position'].max()+10000000, 10000000)]
chrom1_regions
```

```python
bases_chr1.update({'Position':chrom1_regions})
bases_chr1 = pd.DataFrame(bases_chr1)
bases_chr1
```

```python
snp_second_chrom = snp_variants[snp_variants['Chromosome']==chromosome[1]]
bases_chr2 = {'A':[0], 'G':[0], 'C':[0], 'T':[0]}
for i in range(0,snp_second_chrom['Position'].max(), 10000000):
    win_variants = snp_second_chrom[(snp_second_chrom['Position'] >= i) & (snp_second_chrom['Position'] <= i + 10000000)]['Reference']
    for base in 'AGCT':
        bases_chr2[base].append(win_variants.value_counts()[base]/win_variants.value_counts().sum())
```

```python
chrom2_regions = [i for i in range(0,snp_second_chrom['Position'].max()+10000000, 10000000)]
chrom2_regions
```

```python
bases_chr2.update({'Position':chrom2_regions})
bases_chr2 = pd.DataFrame(bases_chr2)
bases_chr2
```

```python
snp_third_chrom = snp_variants[snp_variants['Chromosome']==chromosome[2]]
bases_chr3 = {'A':[0], 'G':[0], 'C':[0], 'T':[0]}
for i in range(0,snp_third_chrom['Position'].max(), 10000000):
    win_variants = snp_third_chrom[(snp_third_chrom['Position'] >= i) & (snp_third_chrom['Position'] <= i+10000000)]['Reference']
    for base in 'AGCT':
        bases_chr3[base].append(win_variants.value_counts()[base]/win_variants.value_counts().sum())
```

```python
chrom3_regions = [i for i in range(0,snp_third_chrom['Position'].max()+10000000, 10000000)]
chrom3_regions
```

```python
bases_chr3.update({'Position':chrom3_regions})
bases_chr3 = pd.DataFrame(bases_chr3)
bases_chr3
```

```python
fig, ax = plt.subplots(3,1)

for base in 'ACGT':
    ax[0].plot(bases_chr1['Position'],bases_chr1[base],linewidth=3,label=base,alpha=0.5)
ax[0].set_xlabel('Position in the chromosome (million base pairs)')
ax[0].set_ylabel('Frequency')
ax[0].legend()

for base in 'ACGT':
    ax[1].plot(bases_chr2['Position'],bases_chr2[base],linewidth=3,label=base,alpha=0.5)
ax[1].set_xlabel('Position in the chromosome (million base pairs)')
ax[1].set_ylabel('Frequency')
ax[1].legend()

for base in 'ACGT':
    ax[2].plot(bases_chr3['Position'],bases_chr3[base],linewidth=3,label=base,alpha=0.5)
ax[2].set_xlabel('Position in the chromosome (million base pairs)')
ax[2].set_ylabel('Frequency')
ax[2].legend()

fig.savefig('base_frequency.pdf')
```

In the case of the first chromosome include annotation text. In particular add to the plot the text `Position of  interest` at a suitable location and point using an arrow to the location where the frequency of the G base starts declining.

```python
fig, ax = plt.subplots()

for base in 'ACGT':
    ax.plot(bases_chr1['Position'],bases_chr1[base],linewidth=3,label=base,alpha=0.5)
ax.set_xlabel('Position in the chromosome (million base pairs)')
ax.set_ylabel('Frequency')
ax.legend()
ax.annotate('Position of interest', xy=(4e07,0.33), 
            xytext=(3.5e07,0.02),arrowprops={})
```
