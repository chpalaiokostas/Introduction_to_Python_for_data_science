### Exercise 1

We will work with the `tilapia_pheno.txt` file.

* Sort the dataframe in terms of `Weight_final` and `Weight_initial` from higher to lower.
* Sort the dataframe based on the difference between `Weight_final` and `Weight_initial` from lower to higher.
* Calculate how many animals are in each Pond of each Location. Also calculate the mean `Weight_final` and `Length_final` and the corresponding standard deviations.
* Calculate how many animals of each line are in each happa of each pond and location. For all numeric columns estimate the mean, max, min and standard deviation.

```python
til_pheno = pd.read_csv('tilapia_pheno.txt', sep="\t", na_values='0')
til_pheno
```

```python
til_pheno.info()
```

```python
til_pheno[['PIT_TAG','Family','Age','Pond','Happa']] = til_pheno[['PIT_TAG','Family','Age','Pond','Happa']].astype(str)
```

```python
til_pheno['PIT_TAG'].duplicated().any()
```

```python
til_pheno.drop_duplicates(subset='PIT_TAG',inplace=True)
```

```python
til_pheno.info()
```

Sort the dataframe in terms of `Weight_final` and `Weight_initial` from higher to lower.

```python
til_pheno.sort_values(['Weight_final','Weight_initial'],ascending=False)
```

Sort the dataframe based on the difference between `Weight_final` and `Weight_initial` from lower to higher.

```python
til_pheno['Weight_diff'] = til_pheno['Weight_final'] - til_pheno['Weight_initial']
```

```python
til_pheno.sort_values('Weight_diff')
```

Calculate how many animals are in each Pond of each Location. Also calculate the mean `Weight_final` and `Length_final` and the corresponding standard deviations.

```python
til_pheno.groupby(['Location','Pond']).size()
```

```python
til_pheno.groupby(['Location','Pond'])[['Weight_final','Length_final']].aggregate(mean=('Weight_final',np.nanmean),std=('Length_final',np.nanstd))
```

Calculate how many animals of each line are in each happa of each pond and site. For all numeric columns estimate the mean, max, min and standard deviation.

```python
til_pheno.groupby(['Location','Pond','Happa']).size()
```

```python
til_pheno.groupby(['Location','Pond','Happa']).aggregate([np.nanmean,np.std,'max','min'])
```

### Exercise 3

We will work with the `variants.txt` file. The dataset contains information about the location of genetic variants across different chromosomes.

* Calculate the number of variants for each chromosome and plot the results in the form of a horizontal bar chart.
* For each chromosome estimate the base distribution of the Reference column. Find out which base is the most common in each chromosome. Do the same for the Alternative column.


```python
variants = pd.read_csv('variants.txt',sep="\t")
variants
```

```python
variants.info()
```

Calculate the number of variants for each chromosome and plot the results in the form of a horizontal bar chart.

```python
variants.groupby('Chromosome').size()
```

```python
variants.groupby('Chromosome').size().plot(kind='barh')
```

For each chromosome estimate the base distribution of the Reference column. Find out which base is the most common in each chromosome. Do the same for the Alternative column.

```python
ref_common = variants.groupby('Chromosome')['Reference'].value_counts()
```

```python
ref_common
```

```python
ref_common_wide = ref_common.unstack()
```

```python
ref_common_wide
```

```python
ref_common_wide.apply(lambda x: (ref_common_wide.columns[x.argmax()], max(x)), axis=1)
```

```python
ref_common.iloc[0::4]
```

```python
alt_common = variants.groupby('Chromosome')['Alternative'].value_counts()
```

```python
alt_common.iloc[0::4]
```

```python
alt_common_wide = alt_common.unstack()
alt_common_wide.apply(lambda x: (alt_common_wide.columns[x.argmax()], max(x)), axis=1)
```

### Exercise 4

We will work with the `brca_data.csv` file. The dataset contains gene expression data related to breast cancer.

* Change the format of the dataframe so that you have as an index the unique values of `PR.Status` column (the shape of your dataframe should be 5x46).
* Change the format of the dataframe so that all genes (columns starting with rs_) are in one columm named as `Gene`.
* Plot the 10 genes with the highest mean expression on positive patients in a bar chart. Do the same for negative patients.
* As above but this time plot the 10 genes with the lowest mean expression in a horizontal bar chart.

```python
brca = pd.read_csv('brca_data.csv')
brca
```

```python
brca.info()
```

```python
brca.dropna(inplace=True)
```

```python
brca.shape
```

Change the format of the dataframe so that you have as an index the unique values of `PR.Status` column (the shape of your dataframe should be 5x46).

```python
brca.pivot_table(index='PR.Status')
```

Change the format of the dataframe so that all genes (columns starting with rs_) are in one columm named as `Gene`.

```python
pd.melt(brca,  id_vars='PR.Status', var_name='Gene', value_name='Expression')
```

Plot the 10 genes with the highest mean expression on positive patients in a bar chart. Do the same for negative patients.

```python
brca_long = pd.melt(brca,  id_vars='PR.Status', var_name='Gene', value_name='Expression')
brca_long[brca_long['PR.Status']=='Positive'].groupby(
    'Gene').agg(np.mean).sort_values(
    'Expression',ascending=False).reset_index().iloc[:10].plot(
    kind='bar',x='Gene',y='Expression')
```

```python
brca_long[brca_long['PR.Status']=='Negative'].groupby(
    'Gene').agg(np.mean).sort_values(
    'Expression',ascending=False).reset_index().iloc[:10].plot(
    kind='bar',x='Gene',y='Expression')
```

As above but this time plot the 10 genes with the lowest expression in a horizontal bar chart.

```python
brca_long[brca_long['PR.Status']=='Positive'].groupby(
    'Gene').agg(np.mean).sort_values(
    'Expression',ascending=True).reset_index().iloc[:10].plot(
    kind='bar',x='Gene',y='Expression')
```

```python
brca_long[brca_long['PR.Status']=='Negative'].groupby(
    'Gene').agg(np.mean).sort_values(
    'Expression',ascending=True).reset_index().iloc[:10].plot(
    kind='bar',x='Gene',y='Expression')
```
